```python
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import re
import numpy as np
import time
```

```python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Chrome/119.0.6045.105"}) # Safari/537.36. Chrome/103.0.0.0
    soup = bts(result.text, "html.parser")
    return soup
```

```python
pages = []
for page in range(2,51):
    pages.append("https://www.arabam.com/ikinci-el/otomobil?currency=TL&minPrice=100000&page="+str(page))
    
pages
```

```python
products = []
for page in pages:
    html = getAndParseURL(page)
    for product in html.findAll("td",{"class":"listing-modelname pr"}):
        products.append("https://www.arabam.com" + product.a.get("href"))
products
```

```python
links = []
for element in products:
    if element.startswith("https://www.arabam.com"):
        links.append(element)
```

```python
features_ = []
for link in links:
    html = getAndParseURL(link)

    try:
        fiyat = html.find("div", {"class": "product-price-container"}).find("div", {"class": "product-price"}).text.strip()
    except:
        fiyat = np.nan
    try:
        ilanno = html.findAll("div", {"class": "property-value"})[0].text
    except:
        ilanno = np.nan
    try:
        tarih = html.findAll("div", {"class": "property-value"})[1].text
    except:
        tarih = np.nan
    try:
        marka = html.findAll("div", {"class": "property-value"})[2].text
    except:
        marka = np.nan
    try:
        seri = html.findAll("div", {"class": "property-value"})[3].text
    except:
        seri = np.nan
    try:
        model = html.findAll("div", {"class": "property-value"})[4].text
    except:
        model = np.nan
    try:
        yil = html.findAll("div", {"class": "property-value"})[5].text
    except:
        yil = np.nan
    try:
        km = html.findAll("div", {"class": "property-value"})[6].text
    except:
        km = np.nan
    try:
        vites = html.findAll("div", {"class": "property-value"})[7].text
    except:
        vites = np.nan
    try:
        yakit = html.findAll("div", {"class": "property-value"})[8].text
    except:
        yakit = np.nan
    try:
        kasa= html.findAll("div", {"class": "property-value"})[9].text
    except:
        kasa = np.nan
    try:
        motorhac = html.findAll("div", {"class": "property-value"})[10].text
    except:
        motorhac = np.nan
    try:
        renk = html.findAll("div", {"class": "property-value"})[11].text
    except:
        renk = np.nan
    try:
        motorpower = html.findAll("div", {"class": "property-value"})[12].text
    except:
        motorpower = np.nan
    try:
        cekis= html.findAll("div", {"class": "property-value"})[13].text
    except:
        cekis = np.nan
    try:
        yakittk = html.findAll("div", {"class": "property-value"})[14].text
    except:
        yakittk = np.nan
    try:
        depo = html.findAll("div", {"class": "property-value"})[15].text
    except:
        depo = np.nan


    
    features_.append([fiyat,ilanno,tarih,marka,seri,model, yil, km, vites, yakit, kasa,motorhac,renk,motorpower,cekis,yakittk,depo])

# Update df_columns to have 9 elements
df_columns = ["Fiyat","lan No","lan Tarihi","Marka","Seri","Model","Yl", "Kilometre", "Vites Tipi", "Yakt Tipi", "Kasa Tipi",
              "Renk", "Motor Hacmi", "Motor Gc", "eki","Ort. Yakt Tketimi","Yakt Deposu"]

df_ = pd.DataFrame.from_records(features_, columns=df_columns)
```
